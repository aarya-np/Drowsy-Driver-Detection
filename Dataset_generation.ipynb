```python
Raw_DIR=r'C:\Users\memas\OneDrive\Desktop\Major Pro\Dataset\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4]=='0':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\memas\OneDrive\Desktop\Major Pro\Prepared_Data\Close Eyes')
            
        elif i.split('_')[4]=='1':
             shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\memas\OneDrive\Desktop\Major Pro\Prepared_Data\Open Eyes')
```
